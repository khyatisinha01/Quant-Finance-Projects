```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Download Netflix stock prices (Close column, since Adj Close was removed by default)
data = yf.download("NFLX", start="2020-01-01", end="2024-01-01")['Close']

# Calculate daily returns
returns = data.pct_change().dropna()

# Parameters
mu = returns.mean()        # average daily return
sigma = returns.std()      # daily volatility
days = 252                 # trading days in a year
simulations = 1000         # number of Monte Carlo simulations
S0 = data.iloc[-1]         # last available stock price

# Run Monte Carlo simulation
simulated_prices = []
for i in range(simulations):
    prices = [S0]
    for d in range(days):
        shock = np.random.normal(mu, sigma)   # random shock
        prices.append(prices[-1] * (1 + shock))
    simulated_prices.append(prices)

# Plot results
plt.figure(figsize=(10,5))
for sim in simulated_prices[:50]:   # plot first 50 paths for clarity
    plt.plot(sim, alpha=0.2, color="blue")

plt.title("Monte Carlo Simulation of Netflix Stock Price (1 Year)")
plt.xlabel("Days")
plt.ylabel("Price (USD)")
plt.show()
```
